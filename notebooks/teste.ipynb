```python
import psycopg2

# Conexão com o banco
conn = psycopg2.connect(
    host="localhost",       # porque você expôs a porta 5432
    port="5432",
    database="mydatabase",
    user="postgres",
    password="postgres"
)

# Criar cursor
cur = conn.cursor()

# Criar tabela de exemplo
cur.execute("""
    CREATE TABLE IF NOT EXISTS clientes (
        id SERIAL PRIMARY KEY,
        nome VARCHAR(100),
        email VARCHAR(100)
    )
""")

# Inserir dados
cur.execute("""
    INSERT INTO clientes (nome, email)
    VALUES (%s, %s)
""", ("José", "jnelior@gmail.com"))

# Confirmar transação
conn.commit()

# Consultar dados
cur.execute("SELECT * FROM clientes")
rows = cur.fetchall()
for row in rows:
    print(row)

# Fechar conexão
cur.close()
conn.close()
```

```python
import pandas as pd
from sqlalchemy import create_engine, VARCHAR

# Conexão
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/mydatabase")

# DataFrame de exemplo
df = pd.DataFrame({
    "nome": ["José", "Maria", "Carlos"],
    "email": ["jose@email.com", "maria@email.com", "carlos@email.com"]
})

# Subir para o banco
df.to_sql(
    name="clientes",
    con=engine,
    schema="public",
    if_exists="replace",
    index=False,
    dtype={"nome": VARCHAR(100), "email": VARCHAR(100)},
    method="multi"
)
```
